```python
import time
import re
import pandas as pd

class Crawler:
    def __init__(self, urls, path):
        self.driver = webdriver.Chrome(path)
        self.urls = urls
        self.dataframe=None
        self.erridx=[]
        self.title_list=[]
        self.author_list=[]
        self.date_list=[]
        self.text_list=[]
        self.buyer_list=[]
        self.isbn_list=[]
        self.category_list=[]
        self.publish_list=[]
    
    def crawl(self):
        col=['name','ISBN','author','date','story','women10','men10','women20','men20','women30','men30','women40','men40','women50','men50','women60','men60','category','publish']
        for i in range (12):
            self.buyer_list.append([])
        for url in self.urls:
            tmp_vars=[]
            self.driver.get(url)
            self.driver.implicitly_wait(3)
            self.title_list.append(self.driver.find_element_by_class_name('Ere_bo_title').text)
            tl=self.driver.find_elements_by_class_name('Ere_sub2_title')
            self.author_list.append(tl[1].text)
            if '원제' in tl[len(tl)-1].text:
                self.publish_list.append(tl[len(tl)-2].text)
            else:
                self.publish_list.append(tl[len(tl)-1].text)
            self.category_list.append('소설/시/희곡')
            day=tl[0].text
            for i in range(1,len(tl)):
                day=day.replace(tl[i].text,"")
            day=day.replace(' (지은이)','').replace(' (옮긴이)','').replace(' (그림)','').replace(',','')
            self.date_list.append(day)
            tmp_isbn = self.driver.find_element_by_class_name('conts_info_list1').find_elements_by_tag_name('li')
            self.isbn_list.append(tmp_isbn[len(tmp_isbn)-1].text.replace('ISBN : ',''))
            self.scroll()
            for i in range(1):
                try:
                    self.driver.find_element_by_xpath('//div[@id="Underline3_more"]/a').click()
                    time.sleep(1.5)
                except:
                    elem = self.driver.find_element_by_tag_name("body")
                    pageup = 1
                    while pageup < 2:
                        elem.send_keys(Keys.PAGE_UP)
                        time.sleep(0.1)
                        pageup += 1
                    for i in range(10):
                        try:
                            self.driver.find_element_by_xpath('//div[@id="Underline3_more"]/a').click()
                            time.sleep(1.1)
                        except:
                            break
            text=''
            li = self.driver.find_elements_by_xpath("//div[@id='Underline3Updates']//div[@class='Rconts2']")
            if len(li)==0:
                ii=self.driver.find_elements_by_xpath("//div[@class='Ere_prod_mconts_box']")
                for i in ii:
                    try:
                        if i.find_element_by_class_name('Ere_prod_mconts_LS').text == '책소개':
                            text = self.dore(i.find_element_by_class_name('Ere_prod_mconts_R').text)
                            break
                    except:
                        None
            else:
                cnt=1
                for i in li:
                    try:
                        if len(self.driver.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"']").text) >= 25:
                            try:
                                i.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"']").click()
                                time.sleep(0.01)
                                text += self.dore(i.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"_more']").text[:-5])
                            except:
                                text += self.dore(i.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"']").text[:-5])
                            else:
                                None
                        else:
                            None
                    except:
                        None
                    cnt+=1
            
            self.text_list.append(text)
            
            buyerli = self.driver.find_elements_by_xpath("//div[@class='buyer_area_a']//div[@class='per']")
            
            for i in range(len(buyerli)):
                self.buyer_list[i].append((buyerli[i].text).replace('%',''))
        
        mylist = {col[0]:self.title_list,col[1]:self.isbn_list,col[2]:self.author_list,col[3]:self.date_list,col[4]:self.text_list,col[5]:self.buyer_list[0],col[6]:self.buyer_list[1],col[7]:self.buyer_list[2],col[8]:self.buyer_list[3],col[9]:self.buyer_list[4],col[10]:self.buyer_list[5],col[11]:self.buyer_list[6],col[12]:self.buyer_list[7],col[13]:self.buyer_list[8],col[14]:self.buyer_list[9],col[15]:self.buyer_list[10],col[16]:self.buyer_list[11],col[17]:self.category_list,col[18]:self.publish_list}
        self.dataframe = pd.DataFrame({ key:pd.Series(value) for key, value in mylist.items() })
    
    def scroll(self):
        for i in range(1,11):
            self.driver.execute_script("window.scrollTo(document.body.scrollHeight/11*"+str(i-1)+", document.body.scrollHeight/11*"+str(i)+");")
            time.sleep(0.5)
            
    def dore(self,text):
        return re.sub('[^a-zA-Z|ㄱ-ㅣ가-힣|0-9]+',' ', text)
        
    def get_df(self):
        return self.dataframe
    
```

```python
crr = Crawler(url_list[0:301],"chromedriver")
```

```python
crr.crawl()
```
