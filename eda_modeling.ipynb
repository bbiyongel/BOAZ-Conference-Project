```python
data = data.dropna()
```

```python
sentence = data['story'].apply(lambda x : re.sub('[^a-zA-Z|ㄱ-ㅣ가-힣|. ]+','', x))
```

```python
sentence
```

```python
from konlpy.tag import Komoran
komo = Komoran()
```

# 문장 나눈거

```python
ts=[]
```

```python
for tag,ss in zip(data['name'],sentence):
    tmp=ss.split('.')
    ttmp=[]
    for i in tmp:
        if(len(re.sub('[^a-zA-Z|ㄱ-ㅣ가-힣]+','', i))!=0):
            ttmp.append(i)
    for sss in ttmp:
        ts.append([tag,komo.nouns(sss)])
```

```python
ts=ts*20
```

```python
from gensim.models.doc2vec import TaggedDocument

class Doc2VecCorpus:
    def __init__(self, ts):
        self.ts = ts
    def __iter__(self):
        for name,sentence in ts:
            if(len(sentence)>0):
                yield TaggedDocument(
                    words = sentence,
                    tags = [name])
```

```python
doc2vec_corpus = Doc2VecCorpus(ts)
```

```python
doc2vec_corpus = Doc2VecCorpus(ts)
from gensim.models import Doc2Vec

doc2vec_model = Doc2Vec(doc2vec_corpus)
```

```python
for idx, doctag in sorted(doc2vec_model.docvecs.doctags.items(), key=lambda x:x[1].offset):
    print(idx, doctag)
```

```python
doc2vec_model.docvecs.most_similar('백설 공주 살인 사건')
```

```python
doc2vec_model.save('doc2vec_tt.h5')
```
