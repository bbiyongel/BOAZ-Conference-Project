```python
import time
import re
import pandas as pd

class Crawler:
    def __init__(self,urls,path):
        self.driver = webdriver.Chrome(path)
        self.urls = urls
        self.dataframe=None
        self.title_list=[]
        self.summary_list=[]
    
    def crawl(self):
        col=['name','summary']
        for url in self.urls:
            tmp_vars=[]
            self.driver.get(url)
            self.driver.implicitly_wait(3)
            self.title_list.append(self.driver.find_element_by_class_name('Ere_bo_title').text)
            tl = self.driver.find_elements_by_class_name('Ere_sub2_title')
            self.scroll()
            for i in range(1):
                try:
                    self.driver.find_element_by_xpath('//*[@id="div_PublisherDesc_Short"]/a[2]').click()
                    time.sleep(1.5)
                except:
                    elem = self.driver.find_element_by_tag_name("body")
                    pageup = 1
                    while pageup < 2:
                        elem.send_keys(Keys.PAGE_UP)
                        time.sleep(0.1)
                        pageup += 1
                    for i in range(10):
                        try:
                            self.driver.find_element_by_xpath('//*[@id="div_PublisherDesc_Short"]/a[2]').click()
                            time.sleep(1.1)
                        except:
                            break
            
            text=''
            ii=self.driver.find_elements_by_xpath("//div[@class='Ere_prod_mconts_box']")
            for i in ii:
                try:
                    if i.find_element_by_class_name('Ere_prod_mconts_LS').text == '출판사 제공 책소개':
                        text = self.dore(i.find_element_by_class_name('Ere_prod_mconts_R').text[:-2])
                    elif i.find_element_by_class_name('Ere_prod_mconts_LS').text == '주간 편집 회의':
                        text = self.dore(i.find_element_by_class_name('Ere_prod_mconts_R').text)
                    else:
                        None
                except:
                    None
            self.summary_list.append(text)

            
            mylist = {col[0]:self.title_list,col[1]:self.summary_list}
            self.dataframe = pd.DataFrame({ key:pd.Series(value) for key, value in mylist.items() })
            
    def dore(self,text):
        return re.sub('[^a-zA-Z|ㄱ-ㅣ가-힣|0-9]+',' ', text)
    
    def scroll(self):
        for i in range(1,11):
            self.driver.execute_script("window.scrollTo(document.body.scrollHeight/11*"+str(i-1)+", document.body.scrollHeight/11*"+str(i)+");")
            time.sleep(0.5)
            
    def get_df(self):
        return self.dataframe
```
